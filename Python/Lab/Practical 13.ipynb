# Computing Average marks of student

```python
marks = []
num = 0
while num >= 0:
    num=float(input("ENTER YOUR MARKS,and ENTER '-1' to stop"))
    if num>=0:
        marks.append(num)
print("You have entered : ",marks)

# Computing average marks
total_marks = 0
for x in range(len(marks)):
    total_marks = total_marks + marks[x]
avg_marks=total_marks/len(marks)

print("The average marks of student is : ",avg_marks)
```
