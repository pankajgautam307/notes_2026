1. Lists in Python


Lists are mutable, ordered collections that allow duplicate values.

Common List Methods :

Method	            Description	                                        Example

append(x)	        Adds an element x at the end of the list.	        lst.append(4)

extend(iterable)	Adds all elements from an iterable to the list.	    lst.extend([5, 6])

insert(index, x)	Inserts x at a given index.	                        lst.insert(1, 100)

remove(x)	        Removes the first occurrence of x.	                lst.remove(2)

pop(index)	        Removes and returns element at index.	            lst.pop(1)

index(x)	        Returns the index of x.	                            lst.index(3)

count(x)	        Returns the number of occurrences of x.	            lst.count(2)

reverse()	        Reverses the list in place.	                        lst.reverse()

sort()	            Sorts the list in ascending order.	                lst.sort()

```python
# Example of List methods :

lst = [1, 2, 3,4,3,2,6,5]
print(lst)

# append()
lst.append(7)  #Adds an element 4 at the end of the list.
print(lst) 
```

2. Tuples in Python
Tuples are immutable, ordered collections that allow duplicates.

Common Tuple Methods:
Method	    Description	                        Example
count(x)	Returns occurrences of x in tuple.	tup.count(2)
index(x)	Returns first index of x.	        tup.index(3)

```python
# 2. Tuples in Python
tup = (1, 2, 3, 2, 4)
print("Count of 2:", tup.count(2))  # Number of times 2 appears
print("Index of 3:", tup.index(3))  # First occurrence of 3
print("Length of tuple : ", len(tup))  # Length of tuple :  5
```

3. Strings in Python
Strings are immutable sequences of characters.

Common String Methods:

Method	        Description	                         Example
lower()	        Converts to lowercase.	             'Hello'.lower() → 'hello'
upper()	        Converts to uppercase.	             'hello'.upper() → 'HELLO'
strip()	        Removes leading/trailing whitespace. ' hello '.strip() → 'hello'
replace(a, b)	Replaces a with b.	                 'hello'.replace('l', 'x') → 'hexxo'
split(sep)	    Splits string into list.	        'a,b,c'.split(',') → ['a', 'b', 'c']
find(sub)	    Finds first index of sub.	        'hello'.find('e') → 1

```python
# 3. Strings in Python
s = "  Hello World  "
print("Given    string:", s)
print("Stripped string:", s.strip())  # Removes spaces
print("Lowercase:", s.lower())  # Converts to lowercase
print("Uppercase:", s.upper())  # Converts to uppercase
print("Replaced string:", s.replace("Hello", "Hi"))  # Replaces words
print("Split string:", s.split())  # Splits into list
print("Finding substring:", s.find("wo"))  # Returns index of substring
```

5. Sets in Python
Sets are unordered, mutable, and store unique elements.

Common Set Methods:
Method	            Description	                                Example
add(x)	            Adds x to set.	                            s.add(4)
remove(x)	        Removes x, raises error if not present.	    s.remove(3)
discard(x)	        Removes x, does nothing if not present.	    s.discard(3)
pop()	            Removes and returns a random element.	    s.pop()
clear()	            Removes all elements.	                    s.clear()
union(set2)	        Returns new set with elements from both.	s1.union(s2)
intersection(set2)	Returns common elements.	                s1.intersection(s2)
difference(set2)	Returns elements in s1 but not in s2.	    s1.difference(s2)

```python
# 5. Sets in Python
s1 = {1, 2, 3}
s2 = {3, 4, 5}


s1.add(6)  # Adds an element
print(s1)
s1.remove(2)  # Removes an element
print(s1)

print("Union:", s1.union(s2))  # Combines sets
print("Intersection:", s1.intersection(s2))  # Common elements
print("Difference:", s1.difference(s2))  # Elements only in s1'
```

6. Dictionaries in Python
Dictionaries are mutable collections of key-value pairs.

Common Dictionary Methods:

Method	            Description	                            Example

keys()	            Returns all keys.	                    d.keys()
values()	        Returns all values.	                    d.values()
items()	            Returns key-value pairs.	            d.items()
get(key, default)	Returns value for key, else default.	d.get('age', 25)
update(dict2)	    Merges dict2 into d.	                d.update({'city': 'NY'})
pop(key)	        Removes key and returns value.	        d.pop('age')
clear()	            Removes all items.	                    d.clear()

```python
# 6. Dictionaries in Python
d = {'name': 'Alice', 'age': 25}
d.update({'city': 'NY','country': "USA"})  # Adds a new key-value pair
print("Dictionary keys:", d.keys())  # Gets all keys
print("Dictionary values:", d.values())  # Gets all values
print("Dictionary items:", d.items())  # Gets key-value pairs
print("Value for 'name':", d.get('age','Pankaj'))  # Gets value for 'name'
x = d.pop('age')
print(x)
print(d)
```
