```python
marks = int(input("Enter your marks"))
if(marks>90):
    print("A+ Grade")
elif(marks>75):
    print("A Grade")
elif(marks>60):
    print("B Grade")
elif(marks>50):
    print("C Grade")
else:
    print("D Grade")


```
